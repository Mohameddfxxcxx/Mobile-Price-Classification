# Training and Testing

In the grand tale of data science, the Train/Test divide marks a crucial chapter, where the dataset is split asunder into two realms: üöÇüîç

The Training Kingdom, where our model learns the secrets of the data, encompassing 80% of its vast expanse. üìöüîÆ

The Testing Dominion, a realm of trials and tribulations, embracing the remaining 20% to put our model's newfound knowledge to the test. üß™üè∞

Thus, our journey unfolds, with the Training Kingdom nurturing the growth of our model's wisdom, while the Testing Dominion stands as the ultimate crucible, where its true strength shall be revealed! üí™üî•

```python
X = train.drop('price_range',axis=1)  # Creating a feature matrix X by removing the 'price_range' column from the training dataset
y = train['price_range']  # Creating a target vector y containing the 'price_range' column values from the training dataset
```

```python
from sklearn.model_selection import train_test_split  # Importing the train_test_split function from the sklearn library for splitting the dataset into training and testing sets

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=100, test_size=0.2)  # Splitting the feature matrix X and target vector y into training and testing sets, with 80% for training and 20% for testing, using a random state for reproducibility
print(X_train.shape, X_test.shape)  # Printing the shapes of the training and testing sets to confirm the split
```

# Model Build


In the intricate dance of algorithm selection, the data reigns supreme, guiding our path toward the perfect match! üõ§Ô∏èüí° Depending on the unique tapestry of our data and the enigma we seek to solve, we embark on a quest to choose the most fitting algorithm, forging a bond between data and model that shall stand the test of time! ‚öîÔ∏èüìä

Enter the humble decision tree: a beacon of clarity in the fog of complexity! üå≥üîç As a non-parametric wonder, it beckons both classification and regression tasks with open arms, its hierarchical structure unfolding like a map to the stars! üåå‚ú® With its roots firmly planted in the soil of data, each node becomes a gatekeeper to knowledge, each branch a path to understanding, and each leaf a testament to wisdom gained! üçÉüîÆ

```python
from sklearn.tree import DecisionTreeClassifier  # Importing the DecisionTreeClassifier class from the sklearn library

dt = DecisionTreeClassifier()  # Creating an instance of the DecisionTreeClassifier class

dt.fit(X_train, y_train)  # Training the decision tree classifier on the training data (X_train and y_train)

y_pred_dt = dt.predict(X_test)  # Making predictions on the test data using the trained decision tree classifier
```

In the enchanted forest of machine learning, the Random Forest stands tall as a majestic guardian, comprised of a legion of decision trees united in purpose! üå≥üå≤ With the power of bagging and the whimsy of feature randomness, it weaves a tapestry of diversity in its leafy canopy, fending off the specter of overfitting and beckoning accuracy to its side! üé©üîÆ As each tree whispers its own tale, the forest as a whole becomes a sanctuary of knowledge and a fortress of wisdom, ready to guide us through the wilderness of data! üèûÔ∏èüåü

```python
from sklearn.ensemble import RandomForestClassifier  # Importing the RandomForestClassifier class from the sklearn library

rf = RandomForestClassifier()  # Creating an instance of the RandomForestClassifier class

rf.fit(X_train, y_train)  # Training the random forest classifier on the training data (X_train and y_train)

y_pred_rf = rf.predict(X_test)  # Making predictions on the test data using the trained random forest classifier
```

In the realm of machine learning, behold the mighty XGBoost! üåüüöÄ This champion among algorithms is no mere tree but a towering colossus of extreme gradient boosted trees, where each tree stands as a sentinel, correcting the errors of its predecessors and paving the way to enlightenment! üå≥üî• With each layer added, the ensemble grows stronger, weaving a tapestry of resilience and adaptability that propels us to new heights of predictive power! üé¢üí• So let us harness the energy of XGBoost, riding the waves of gradient descent to conquer the peaks of accuracy and forge a path to victory! üèÜüåà

```python
!pip install xgboost
from xgboost import XGBClassifier  # Importing the XGBClassifier class from the xgboost library

xgb = XGBClassifier()  # Creating an instance of the XGBClassifier class

xgb.fit(X_train, y_train)  # Training the XGBoost classifier on the training data (X_train and y_train)

y_pred_xgb = xgb.predict(X_test)  # Making predictions on the test data using the trained XGBoost classifier
```

# Model Evaluation 

In the grand arena of model evaluation, Accuracy emerges as a shining beacon of success, illuminating the path to greatness! üåüüîç This noble metric measures the harmony between our model's predictions and the ground truth, revealing the proportion of triumphs amidst the trials of evaluation! üèÜüí° As a steadfast companion on our journey, Accuracy stands ready to assess the prowess of our classification model, especially in the serene waters where classes are balanced and victory lies within reach! üöÄüéØ Let us embrace this metric with open arms, as we strive for excellence in the realm of machine learning! üåüüìä

```python
from sklearn.metrics import confusion_matrix  # Importing the confusion_matrix function from the sklearn.metrics module for evaluating the classification model's performance

from sklearn.metrics import accuracy_score  # Importing the accuracy_score function from the sklearn.metrics module for calculating the accuracy of the classification model
```

```python
dt_acc = accuracy_score(y_pred_dt, y_test)  # Calculating the accuracy score of the Decision Tree Classifier Algorithm by comparing the predicted labels (y_pred_dt) with the actual labels from the test set (y_test)

print("Accuracy score of Decision Tree Classifier Algorithm is: ", dt_acc)  # Printing the accuracy score of the Decision Tree Classifier Algorithm

print(confusion_matrix(y_pred_dt,y_test))  # Printing the confusion matrix to evaluate the performance of the Decision Tree Classifier Algorithm
```

```python
rf_acc = accuracy_score(y_pred_rf, y_test)  # Calculating the accuracy score of the Random Forest Classifier Algorithm by comparing the predicted labels (y_pred_rf) with the actual labels from the test set (y_test)

print("Accuracy score of Random Forest Classifier Algorithm is: ", rf_acc)  # Printing the accuracy score of the Random Forest Classifier Algorithm

print(confusion_matrix(y_pred_rf,y_test))  # Printing the confusion matrix to evaluate the performance of the Random Forest Classifier Algorithm
```

```python
xgb_acc = accuracy_score(y_pred_xgb, y_test)  # Calculating the accuracy score of the XGBoost Classifier Algorithm by comparing the predicted labels (y_pred_xgb) with the actual labels from the test set (y_test)

print("Accuracy score of XGBoost Classifier Algorithm is: ", xgb_acc)  # Printing the accuracy score of the XGBoost Classifier Algorithm

print(confusion_matrix(y_pred_xgb,y_test))  # Printing the confusion matrix to evaluate the performance of the XGBoost Classifier Algorithm
```

```python
plt.bar(x=['dtc','rfc','xgb'],height=[dt_acc,rf_acc,xgb_acc])  # Creating a bar plot to visualize the accuracy scores of different algorithms
plt.xlabel("Algorithms")  # Adding label for the x-axis
plt.ylabel("Accuracy Score")  # Adding label for the y-axis
plt.show()  # Displaying the bar plot
```
